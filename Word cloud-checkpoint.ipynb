## Word Cloud 
Un word cloud (nuage de mots) est une repr√©sentation visuelle d‚Äôun texte o√π la taille des mots refl√®te leur fr√©quence. Plus un mot appara√Æt souvent dans le texte (comme des avis clients), plus il est affich√© en gros.

## üéØ Utilit√© :
* Identifier rapidement les mots cl√©s et th√®mes dominants
* R√©sumer des avis, commentaires, ou r√©sultats d‚Äôenqu√™te
* Visualisation simple et impactante

### Remplacer les emojis par des commentaires

```python
# √âtape 1 : Stocker le nombre de commentaires fran√ßais AVANT
nb_fr_avant = len(df[df['Langue'] == 'fr'])

# √âtape 2 : D√©finir la fonction si ce n'est pas encore fait
def remplacer_emoji(avis):
    if any(emoji in avis for emoji in ['ü•∞', '‚ù§Ô∏è', 'ü§ç', 'üòÉ', 'üòç', '‚ô•Ô∏è', 'üòª']):
        return 'positif'
    elif any(emoji in avis for emoji in ['ü§¨', 'üëéüèª']):
        return 'n√©gatif'
    elif any(emoji in avis for emoji in ['üëç']):
        return 'neutre'
    else:
        return 'emoji'

# √âtape 3 : Identifier les lignes √† modifier
indices = df[(df['Langue'] == 'unknown') & 
             (df['Avis Client'].apply(remplacer_emoji).isin(['positif', 'n√©gatif', 'neutre']))].index

# √âtape 4 : Modifier les langues pour ces lignes
df.loc[indices, 'Langue'] = 'fr'

# √âtape 5 : Nombre d'avis fran√ßais APR√àS
nb_fr_apres = len(df[df['Langue'] == 'fr'])

# √âtape 6 : Affichage du r√©sultat
print(f"Nombre de commentaires fran√ßais avant : {nb_fr_avant}")
print(f"Nombre de commentaires fran√ßais apr√®s  : {nb_fr_apres}")
print(f"Commentaires ajout√©s √† 'fr' : {nb_fr_apres - nb_fr_avant}")
```

```python
from collections import Counter
import re
#Counter est un outil de comptage int√©gr√© √† Python (dans le module collections) qui permet de :
    #Compter combien de fois chaque √©l√©ment (ex: mot, lettre, chiffre) appara√Æt dans une liste
    #Trouver les √©l√©ments les plus fr√©quents
```

```python
#Modifier la mot merci

def nettoyer_merci(texte):
    return re.sub(r'merci+i+!?', 'Merci', str(texte), flags=re.IGNORECASE)

# Appliquer sur la colonne
df['Avis Client'] = df['Avis Client'].apply(nettoyer_merci)
```

```python
# D√©finir les langues pr√©vues
langues_attendues = ['fr', 'en', 'ar']

# Filtrer les commentaires non pr√©vus (donc probablement en tunisien)
df_tn = df[~df['Langue'].isin(langues_attendues)].copy()
```

```python
# Concat√©ner tous les commentaires
textes_tn = ' '.join(df_tn['Avis Client'].dropna().astype(str))

# Nettoyer le texte (enlever ponctuations et mettre en minuscule)
textes_tn = re.sub(r'[^a-zA-Z0-9\u0600-\u06FF\s]', '', textes_tn.lower())

# D√©couper en mots
mots = textes_tn.split()
```

```python
def nettoyer_texte_emoji(texte):
    texte = re.sub(r'[^\w\s]', '', texte)  # Enl√®ve ponctuation et emojis
    return texte.strip().lower()

def detect_lang_corrige(text):
    try:
        texte_nettoye = nettoyer_texte_emoji(text)
        return detect(texte_nettoye) if texte_nettoye else 'unknown'
    except:
        return 'unknown'

# Appliquer sur ta colonne
df['Langue'] = df['Avis Client'].apply(detect_lang_corrige)
```

```python
def ajuster_langue(commentaire):
    commentaire_str = str(commentaire).lower()
    mots_cles_fr = ["je t'aime", "merci", "j'aime", "r√©seau", "connexion"]
    if any(mot in commentaire_str for mot in mots_cles_fr):
        return 'fr'
    else:
        return detect_lang_corrige(commentaire)
```

```python
def remplacement_personnalise(texte):
    texte = str(texte).lower().strip()

    # Dictionnaire de remplacements directs
    remplacements = {
        'soooo baaad': 'so bad',
        'gooood': 'good',
        'baaaad': 'bad',
        'thank u <3': 'thank you',
        'thank u': 'thank you'
    }

    # V√©rifie si le texte correspond exactement √† une cl√©
    if texte in remplacements:
        return remplacements[texte]

    return texte  # sinon, retourne le texte inchang√©
```

```python
# √âtape 1 : D√©tection simple de langue bas√©e sur les mots apr√®s nettoyage
def detect_langue_simple(avis):
    avis_lower = avis.lower()

    # Mots indicateurs de fran√ßais (apr√®s nettoyage)
    mots_fr = ['null', "c'est", 'merci', 'reseau', 'super', 'beaucoup']
    # Mots indicateurs d‚Äôanglais
    mots_en = ['nice', 'wifi', 'modem']

    # V√©rification
    if any(mot in avis_lower for mot in mots_fr):
        return 'fr'
    elif any(mot in avis_lower for mot in mots_en):
        return 'en'
    else:
        return 'ro'  # Ne pas modifier si on ne sait pas

# √âtape 2 : Appliquer uniquement sur les lignes qui sont encore en 'ro'
df.loc[df['Langue'] == 'ro', 'Langue'] = df[df['Langue'] == 'ro']['Avis Client'].apply(detect_langue_simple)

# √âtape 3 : V√©rification du r√©sultat
print(df[df['Langue'].isin(['fr', 'en'])][['Avis Client', 'Langue']])
```

```python
# Mots-cl√©s qui indiquent du fran√ßais
mots_fr = ['merci', 'application', 'alternative', 'bonne', 'mediocre']

def detect_fr_par_mots(texte):
    texte_lower = texte.lower()
    return any(mot in texte_lower for mot in mots_fr)

# Appliquer la correction uniquement sur les lignes o√π la langue est 'it' (italien)
mask = (df['Langue'] == 'it') & (df['Avis Client'].apply(detect_fr_par_mots))
df.loc[mask, 'Langue'] = 'fr'

# V√©rifier les r√©sultats
print(df[mask][['Avis Client', 'Langue']])
```

## Cr√©er un Word Cloud en arabe

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
```

```python
# Filtrer uniquement les textes en arabe
arabic_texts = df.loc[df['Langue'] == 'ar', 'Avis Client']
```

```python
# Concat√©ner tous les textes en un seul grand texte
full_text = " ".join(arabic_texts)
```

```python
# Reshape pour la bonne forme des lettres arabes
reshaped_text = arabic_reshaper.reshape(full_text)
```

```python
# Corriger la direction d'√©criture (RTL)
bidi_text = get_display(reshaped_text)
```

```python
# Cr√©er le WordCloud (tu peux ajouter une police arabe si n√©cessaire)
wordcloud = WordCloud(
    font_path='arial.ttf',  # Remplacer par un chemin vers une police compatible arabe (ex: Arial, Amiri, etc.)
    background_color='white',
    width=800,
    height=400
).generate(bidi_text)

# Afficher
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Word cloud multilingue 

```python
# Filtrer les textes en arabe, anglais et fran√ßais
arabic_texts = df.loc[df['Langue'] == 'ar', 'Avis Client']
english_texts = df.loc[df['Langue'] == 'en', 'Avis Client']
french_texts = df.loc[df['Langue'] == 'fr', 'Avis Client']

# Concat√©ner tous les textes par langue
full_arabic = " ".join(arabic_texts)
full_english = " ".join(english_texts)
full_french = " ".join(french_texts)
```

```python
# Reshape et bidi uniquement pour l'arabe
reshaped_arabic = arabic_reshaper.reshape(full_arabic)
bidi_arabic = get_display(reshaped_arabic)

# Combiner tous les textes en un seul texte final
combined_text = f"{bidi_arabic} {full_english} {full_french}"
```

```python
# Cr√©er le word cloud
wordcloud = WordCloud(
    font_path='arial.ttf',  # Police compatible arabe 
    background_color='white',
    width=900,
    height=450,
    max_words=300
).generate(combined_text)
```

```python
# Afficher
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
