## Word Cloud 
Un word cloud (nuage de mots) est une reprÃ©sentation visuelle dâ€™un texte oÃ¹ la taille des mots reflÃ¨te leur frÃ©quence. Plus un mot apparaÃ®t souvent dans le texte (comme des avis clients), plus il est affichÃ© en gros.

## ğŸ¯ UtilitÃ© :
* Identifier rapidement les mots clÃ©s et thÃ¨mes dominants
* RÃ©sumer des avis, commentaires, ou rÃ©sultats dâ€™enquÃªte
* Visualisation simple et impactante

```python
from collections import Counter
import re
#Counter est un outil de comptage intÃ©grÃ© Ã  Python (dans le module collections) qui permet de :
    #Compter combien de fois chaque Ã©lÃ©ment (ex: mot, lettre, chiffre) apparaÃ®t dans une liste
    #Trouver les Ã©lÃ©ments les plus frÃ©quents
```

```python
#Modifier la mot merci

def nettoyer_merci(texte):
    return re.sub(r'merci+i+!?', 'Merci', str(texte), flags=re.IGNORECASE)

# Appliquer sur la colonne
df['Avis Client'] = df['Avis Client'].apply(nettoyer_merci)
```

```python
# DÃ©finir les langues prÃ©vues
langues_attendues = ['fr', 'en', 'ar']

# Filtrer les commentaires non prÃ©vus (donc probablement en tunisien)
df_tn = df[~df['Langue'].isin(langues_attendues)].copy()
```

```python
def nettoyer_texte_emoji(texte):
    texte = re.sub(r'[^\w\s]', '', texte)  # EnlÃ¨ve ponctuation et emojis
    return texte.strip().lower()

def detect_lang_corrige(text):
    try:
        texte_nettoye = nettoyer_texte_emoji(text)
        return detect(texte_nettoye) if texte_nettoye else 'unknown'
    except:
        return 'unknown'

# Appliquer sur ta colonne
df['Langue'] = df['Avis Client'].apply(detect_lang_corrige)
```

```python
def ajuster_langue(commentaire):
    commentaire_str = str(commentaire).lower()
    mots_cles_fr = ["je t'aime", "merci", "j'aime", "rÃ©seau", "connexion"]
    if any(mot in commentaire_str for mot in mots_cles_fr):
        return 'fr'
    else:
        return detect_lang_corrige(commentaire)
```

```python
def remplacement_personnalise(texte):
    texte = str(texte).lower().strip()

    # Dictionnaire de remplacements directs
    remplacements = {
        'soooo baaad': 'so bad',
        'gooood': 'good',
        'baaaad': 'bad',
        'thank u <3': 'thank you',
        'thank u': 'thank you'
    }

    # VÃ©rifie si le texte correspond exactement Ã  une clÃ©
    if texte in remplacements:
        return remplacements[texte]

    return texte  # sinon, retourne le texte inchangÃ©
```

```python
# Ã‰tape 1 : DÃ©tection simple de langue basÃ©e sur les mots aprÃ¨s nettoyage
def detect_langue_simple(avis):
    avis_lower = avis.lower()

    # Mots indicateurs de franÃ§ais (aprÃ¨s nettoyage)
    mots_fr = ['null', "c'est", 'merci', 'reseau', 'super', 'beaucoup']
    # Mots indicateurs dâ€™anglais
    mots_en = ['nice', 'wifi', 'modem']

    # VÃ©rification
    if any(mot in avis_lower for mot in mots_fr):
        return 'fr'
    elif any(mot in avis_lower for mot in mots_en):
        return 'en'
    else:
        return 'ro'  # Ne pas modifier si on ne sait pas

# Ã‰tape 2 : Appliquer uniquement sur les lignes qui sont encore en 'ro'
df.loc[df['Langue'] == 'ro', 'Langue'] = df[df['Langue'] == 'ro']['Avis Client'].apply(detect_langue_simple)

# Ã‰tape 3 : VÃ©rification du rÃ©sultat
print(df[df['Langue'].isin(['fr', 'en'])][['Avis Client', 'Langue']])
```

```python
# Mots-clÃ©s qui indiquent du franÃ§ais
mots_fr = ['merci', 'application', 'alternative', 'bonne', 'mediocre']

def detect_fr_par_mots(texte):
    texte_lower = texte.lower()
    return any(mot in texte_lower for mot in mots_fr)

# Appliquer la correction uniquement sur les lignes oÃ¹ la langue est 'it' (italien)
mask = (df['Langue'] == 'it') & (df['Avis Client'].apply(detect_fr_par_mots))
df.loc[mask, 'Langue'] = 'fr'

# VÃ©rifier les rÃ©sultats
print(df[mask][['Avis Client', 'Langue']])
```

## CrÃ©er un Word Cloud en arabe

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
```

```python
# Filtrer uniquement les textes en arabe
arabic_texts = df.loc[df['Langue'] == 'ar', 'Avis Client']
```

```python
# ConcatÃ©ner tous les textes en un seul grand texte
full_text = " ".join(arabic_texts)
```

```python
# Reshape pour la bonne forme des lettres arabes
reshaped_text = arabic_reshaper.reshape(full_text)
```

```python
# Corriger la direction d'Ã©criture (RTL)
bidi_text = get_display(reshaped_text)
```

```python
# CrÃ©er le WordCloud (tu peux ajouter une police arabe si nÃ©cessaire)
wordcloud = WordCloud(
    font_path='arial.ttf',  # Remplacer par un chemin vers une police compatible arabe (ex: Arial, Amiri, etc.)
    background_color='white',
    width=800,
    height=400
).generate(bidi_text)

# Afficher
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Word cloud multilingue 

```python
# Filtrer les textes en arabe, anglais et franÃ§ais
arabic_texts = df.loc[df['Langue'] == 'ar', 'Avis Client']
english_texts = df.loc[df['Langue'] == 'en', 'Avis Client']
french_texts = df.loc[df['Langue'] == 'fr', 'Avis Client']

# ConcatÃ©ner tous les textes par langue
full_arabic = " ".join(arabic_texts)
full_english = " ".join(english_texts)
full_french = " ".join(french_texts)
```

```python
# Reshape et bidi uniquement pour l'arabe
reshaped_arabic = arabic_reshaper.reshape(full_arabic)
bidi_arabic = get_display(reshaped_arabic)

# Combiner tous les textes en un seul texte final
combined_text = f"{bidi_arabic} {full_english} {full_french}"
```

```python
# CrÃ©er le word cloud
wordcloud = WordCloud(
    font_path='arial.ttf',  # Police compatible arabe 
    background_color='white',
    width=900,
    height=450,
    max_words=300
).generate(combined_text)
```

```python
# Afficher
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
